```python
#calculating t1
tot = 0
v0 = 100 # m/s
B = 1 # alpha/mass
g = -9.81
v1 = v0
s = 0
tvalx = []
vvaly = []
deltat = 0.001
while v1 > 0:
    # print(v1)  
    tvalx.append(tot)
    vvaly.append(v1)
    s = s + v1*deltat
    v1 = v0 + (g - B*v0)*deltat
    v0 = v1
    tot = tot + deltat
print(tot)
print(s)

#plotting t1
import matplotlib.pyplot as plt  
x = tvalx 
y = vvaly 
    
plt.plot(x, y) 
plt.xlabel('Time') 
plt.ylabel('Velocity')  
plt.title('Velocity vs Time') 
print(plt.show())
```

```python
#checking s(t1) formula
v0 = 100
print((g*tot+v0)/B)
print(tot)
```
